```python
import pandas as pd
import plotly.express as px
import streamlit as st

car_data = pd.read_csv('vehicles.csv') # lendo os dados
fig = px.histogram(car_data, x="type") # criar um histograma
fig.show() # exibindo
```

```python
car_data = pd.read_csv('vehicles.csv') # lendo os dados
hist_button = st.button('Criar histograma')
```

```python
if hist_button: # se o boto for clicado
    # escrever uma mensagem
    st.write('Criando um histograma para o conjunto de dados de anncios de vendas de carros')
            
    # criar um histograma
    fig = px.histogram(car_data, x="type")
        
    # exibir um grfico Plotly interativo
    st.plotly_chart(fig, use_container_width=True)
```

```python
scatter_button = st.button('Criar Grfico de disperso')

if scatter_button: # se o boto for clicado
    # escrever uma mensagem
    st.write('Criando um grfico de disperso para o conjunto de dados de anncios de vendas de carros')
            
    # criar um histograma
    fig = px.scatter(car_data, x="type", y="price")
        
    # exibir um grfico Plotly interativo
    st.plotly_chart(fig, use_container_width=True)
```
